```python
import pymongo
import psycopg2
import pandas as pd
from sqlalchemy import create_engine
```

```python
client = pymongo.MongoClient("mongodb://localhost:27017")
db = client["wiki_crawled"]
print(db)
```

```python
wiki_collection = db["wikipedia"]
wiki = wiki_collection.find()
list_wiki = []
for w in wiki:
    print(w)
    list_wiki.append(w)
    
```

```python
df = pd.DataFrame(list_wiki)
```

```python
#IMPORT DATA TO POSTGRES
import psycopg2
import pandas as pd
from sqlalchemy import create_engine


conn_string = 'postgres://test1:test1@localhost/news_crawled'

db = create_engine(conn_string)
conn = db.connect()

df.to_sql('Wiki', con=conn, if_exists='replace',
		index=False)
conn = psycopg2.connect(conn_string
						)
conn.autocommit = True
cursor = conn.cursor()

sql1 = '''select * from Wiki;'''
cursor.execute(sql1)
for i in cursor.fetchall():
	print(i)

# conn.commit()
conn.close()
```

```python
#Load data from PostgresSQL
import psycopg2

import pandas as pd

from sqlalchemy import create_engine

 

# Create an engine instance

alchemyEngine  = create_engine('postgres://test1:test1@localhost/news_crawled');

 # Connect to PostgreSQL server

dbConnection    = alchemyEngine.connect();
 

# Read data from PostgreSQL database table and load into a DataFrame instance

dataFrame  = pd.read_sql("select * from \"Wiki\"", dbConnection);

# pd.set_option('display.expand_frame_repr', False);

# Print the DataFrame
print(dataFrame);
tmp = dataFrame
# Close the database connection
dbConnection.close();
```
